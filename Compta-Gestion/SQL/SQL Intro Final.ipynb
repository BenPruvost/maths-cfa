## Création de la base de données Logilivres

```sql
CREATE TABLE Librairies
(
    NumLib INT PRIMARY KEY NOT NULL,
    NomLib TEXT,
    AdresseLib TEXT,
    TelLib TEXT
)
```

```sql
CREATE TABLE Livres
(
    NumLivre INT PRIMARY KEY NOT NULL,
    NomLivre TEXT NOT NULL,
    AuteurLivre TEXT NOT NULL,
    EditeurLivre TEXT,
    PrixLivre NUMERIC NOT NULL
);

CREATE TABLE Commandes
(
    NumCommande INT PRIMARY KEY NOT NULL,
    DateCommande DATE,
    NumLib INT
);

CREATE TABLE CommandeLivre
(
    NumLivre INT NOT NULL,
    NumCommande INT NOT NULL,
    QuantiteLivre INT
);
```

## Jointures de tables

Pour extraire des données provenant de plusieurs tables, il faut d'abord les joindre

On veut lier les commandes et les librairies qui commandent

```sql
SELECT Commandes.NumCommande, Librairies.NomLib, Commandes.DateCommande, Librairies.AdresseLib
FROM Commandes
INNER JOIN Librairies ON Librairies.NumLib = Commandes.NumLib
```

```sql
SELECT Commandes.NumCommande, Librairies.NomLib, Commandes.DateCommande, Librairies.AdresseLib
FROM Commandes
INNER JOIN Librairies ON Librairies.NumLib = Commandes.NumLib
WHERE Commandes.DateCommande >= "2025/10/01"
```

```sql
SELECT CommandeLivre.NumCommande, Livres.NomLivre, Livres.PrixLivre, CommandeLivre.QuantiteLivre
FROM CommandeLivre
JOIN Livres ON Livres.NumLivre = CommandeLivre.NumLivre
```

On peut joindre plus de 2 tables avec une jointure :

```sql
SELECT Commandes.NumCommande, Commandes.DateCommande, Librairies.NomLib,
        Librairies.AdresseLib, Livres.NomLivre, Livres.PrixLivre, CommandeLivre.QuantiteLivre
FROM Commandes
JOIN Librairies ON Librairies.NumLib = Commandes.NumLib
JOIN CommandeLivre ON CommandeLivre.NumCommande = Commandes.NumCommande
JOIN Livres ON Livres.NumLivre = CommandeLivre.NumLivre
```

```sql
SELECT Commandes.NumCommande, Commandes.DateCommande, Librairies.NomLib,
        Librairies.AdresseLib, Livres.NomLivre, Livres.PrixLivre, CommandeLivre.QuantiteLivre
FROM Livres
JOIN CommandeLivre ON CommandeLivre.NumLivre = Livres.NumLivre
JOIN Commandes ON Commandes.NumCommande = CommandeLivre.NumCommande
JOIN Librairies ON Librairies.NumLib = Commandes.NumLib
```

## Fonctions d'agrégation

COUNT(colonne) permet de compter le nombre d'entrées remplies d'une colonne.<br/>
COUNT(\*) permet de compter le nombre d'entrées d'une Table

```sql
SELECT COUNT(*) AS NombreEntrees FROM Livres
```

```sql
SELECT AVG(PrixLivre) AS PrixMoyen FROM Livres
```

```sql
SELECT ROUND(AVG(PrixLivre),2) AS PrixMoyen FROM Livres
```

```sql
SELECT SUM(QuantiteLivre) AS QuantiteLivresVendus
FROM CommandeLivre
```

```sql
SELECT SUM(QuantiteLivre) AS QuantiteLivresVendus
FROM CommandeLivre
WHERE NumCommande <= 2
```

Si je veux le prix total de chaque Commande, il va falloir utiliser GROUP BY NumCommande pour la fonction d'agrégation.

```sql
SELECT CommandeLivre.NumCommande, SUM(Livres.PrixLivre*CommandeLivre.QuantiteLivre) AS PrixFacture
FROM CommandeLivre
JOIN Livres ON Livres.NumLivre = CommandeLivre.NumLivre
GROUP BY CommandeLivre.NumCommande
```

```sql
SELECT SUM(Livres.PrixLivre*CommandeLivre.QuantiteLivre) AS CA
FROM CommandeLivre
JOIN Livres ON Livres.NumLivre = CommandeLivre.NumLivre
```
