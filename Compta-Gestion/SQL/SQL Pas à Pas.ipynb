# Créer et manipuler des tables

## Création des tables Librairies, Livres, Commandes et Commandes_Livres

```sql
CREATE TABLE Librairies
(
    NumLib INT,
    NomLib TEXT,
    AdresseLib TEXT,
    TelLib TEXT
);
```

```sql
CREATE TABLE Livres
(
    NumLivre INT,
    NomLivre TEXT,
    AuteurLivre TEXT,
    EditeurLivre TEXT,
    PrixLivre NUMERIC
);
```

```sql
CREATE TABLE Commandes
(
    NumCommande INT,
    NumCli INT,
    DateCommande DATE
);
CREATE TABLE CommandesLivres
(
    NumCommande INT,
    NumLivre INT,
    QuantiteLivre INT
)
```

# Extraction des données de plusieurs tables

```sql
SELECT Commandes.NumCommande, Librairies.NomLib, Commandes.DateCommande, Librairies.AdresseLib
FROM Commandes
JOIN Librairies ON Commandes.NumCli = Librairies.NumLib
WHERE Commandes.DateCommande < "2025-10-01" ;
```

```sql
SELECT CommandesLivres.NumCommande, Livres.NomLivre, Livres.PrixLivre, CommandesLivres.QuantiteLivre
FROM CommandesLivres
JOIN Livres ON CommandesLivres.NumLivre = Livres.NumLivre
```

## Jointures sur plusieurs tables

```sql
SELECT Commandes.NumCommande, Librairies.NomLib, Commandes.DateCommande, Librairies.AdresseLib, 
        CommandesLivres.NumLivre, CommandesLivres.QuantiteLivre
FROM Commandes
JOIN Librairies ON Commandes.NumCli = Librairies.NumLib
JOIN CommandesLivres ON Commandes.NumCommande = CommandesLivres.NumCommande
WHERE Commandes.DateCommande < "2025-10-01" ;
```

```sql
SELECT Commandes.NumCommande, Librairies.NomLib, Librairies.AdresseLib, Commandes.DateCommande, 
        Livres.NomLivre, Livres.PrixLivre, CommandesLivres.QuantiteLivre
FROM CommandesLivres
JOIN Commandes ON Commandes.NumCommande = CommandesLivres.NumCommande 
JOIN Librairies ON Commandes.NumCli = Librairies.NumLib
JOIN Livres ON CommandesLivres.NumLivre = Livres.NumLivre;
```

# Fonctions d'agrégation

```sql
SELECT COUNT(*) FROM Livres
```

```sql
SELECT AVG(PrixLivre) AS "PrixMoyen"
FROM LIVRES
```

```sql
SELECT ROUND(AVG(PrixLivre),2) AS "PrixMoyen"
FROM LIVRES
```

```sql
SELECT SUM(QuantiteLivre) AS "NumLivresCommandes"
FROM CommandesLivres
WHERE NumCommande < 3;
```

```sql
SELECT NumCommande, Livres.NomLivre, QuantiteLivre, SUM(Livres.PrixLivre*QuantiteLivre) AS "PrixTotal"
FROM CommandesLivres
JOIN Livres ON CommandesLivres.NumLivre = Livres.NumLivre
GROUP BY NumCommande, Livres.NumLivre
```

```sql
SELECT Commandes.NumCommande, Commandes.DateCommande, Librairies.NomLib, Librairies.AdresseLib,
        GROUP_CONCAT(Livres.NomLivre) AS "LivresCommandes", 
        SUM(Livres.PrixLivre*QuantiteLivre) AS "PrixTotal"
FROM CommandesLivres
JOIN Livres ON CommandesLivres.NumLivre = Livres.NumLivre
JOIN Commandes ON Commandes.NumCommande = CommandesLivres.NumCommande
JOIN Librairies ON Librairies.NumLib = Commandes.NumCli
GROUP BY Commandes.NumCommande, Commandes.NumCli
```
