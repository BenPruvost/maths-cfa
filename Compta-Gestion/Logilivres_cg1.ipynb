Ecrire la requête permettant d’afficher le numéro de la commande, les livres commandés, leurs prix et leur quantité :

```sql
SELECT Contient.NumComm, NomLivre, PrixLivre, QuantiteLivre
FROM Contient
JOIN Livre ON Contient.NumLivre = Livre.NumLivre
```

```sql
SELECT Contient.NumComm, NomLivre, PrixLivre, QuantiteLivre
FROM Contient, Livre
WHERE Contient.NumLivre = Livre.NumLivre
```

Ecrire la requête permettant d’afficher le numéro et la date de la commande, le nom et l’adresse de la librairie, les livres commandés ainsi que leurs prix et leur quantité :

```sql
SELECT c.NumComm, DateComm, NomLib, AdresseLib, 
    NomLivre, PrixLivre, QuantiteLivre
FROM Contient c, Livre liv, Librairie lib, Commande com
WHERE c.NumLivre = liv.NumLivre
AND c.NumComm = com.NumComm
AND com.NumLib = lib.NumLib
```
