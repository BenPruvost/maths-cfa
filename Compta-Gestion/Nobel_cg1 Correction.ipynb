a) obtenir les catégories dans lesquelles sont attribuées les prix Nobel 

```sql
SELECT DISTINCT category FROM nobel
```

b) lister par ordre alphabétique ( sans doublon) les lauréats du prix Nobel nés en France ou travaillant pour une organisation Française ;

```sql
SELECT DISTINCT FullName AS "Nom", Category AS "Catégorie", Year AS "Année" 
FROM nobel
WHERE BirthCountry = "France"
OR OrganizationCountry = "France"
ORDER BY Fullname
```

c) lister les années où le "Comité international de la Croix Rouge" a obtenu le prix Nobel

```sql
SELECT Year, Fullname FROM nobel
WHERE FullName LIKE "% Croix Rouge %"
```

d) connaître le nombre de femmes ayant obtenu un prix Nobel 

```sql
SELECT Sex, COUNT(DISTINCT Fullname) AS "Nombre" FROM nobel
GROUP BY Sex
```

```sql
SELECT Sex, COUNT(Fullname) AS "Nombre" FROM nobel
GROUP BY Sex
```

e)  lister les femmes françaises ayant obtenu un prix Nobel depuis 2005 ainsi que la catégorie de ce prix Nobel

```sql
SELECT Fullname, Category FROM Nobel
WHERE Sex = "Female"
AND BirthCountry = "France"
AND Year >= 2005
```

f) lister par âge décroissant les lauréats individuels du prix Nobel qui sont toujours en vie (leur DeathYear est alors NULL)

```sql
SELECT DISTINCT Fullname FROM nobel
WHERE DeathDate IS Null
ORDER BY BirthDate DESC;
```

g) rechercher les lauréats dont le nom contient "Curie"

```sql
SELECT FullName, Motivation FROM Nobel
WHERE FullName LIKE "%Curie%"
```
