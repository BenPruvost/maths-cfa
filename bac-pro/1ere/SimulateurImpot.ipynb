```python
def CalculImpotQF(QF):
    tranches=[0,10000,30000,80000,200000]
    taux=[0,0.15,0.3,0.4,0.5]

    ResteAImposer=QF
    i=0
    ImpotParPart=0

    while ResteAImposer>0:
        if QF > tranches[i+1]:
            ImpotParPart+=(tranches[i+1]-tranches[i])*taux[i]
        else:
            ImpotParPart+=ResteAImposer*taux[i]

        ResteAImposer=QF-tranches[i+1]
        i+=1
        if i==4 and ResteAImposer>0:
            ImpotParPart+=ResteAImposer*taux[i]
            ResteAImposer=0
            
    return ImpotParPart
```

```python
def Simulateur():
    RevenuNetImposable=float(input("Revenu net imposable : "))
    NombreParts=float(input("Nombre de Parts : "))

    QF = RevenuNetImposable/NombreParts

    print()
    print("Le quotient familial est :",int(QF*100)/100)

    ImpotParPart=CalculImpotQF(QF)

    print()
    print("Il faudra payer",ImpotParPart,"euros par part.")
    print()

    ImpotTotal = ImpotParPart*NombreParts
    print("Le montant annuel des impôts sera de",int(ImpotTotal*100)/100)
    print()

    TauxMoyen = ImpotTotal/RevenuNetImposable
    print("Le taux moyen d'imposition est de",int(TauxMoyen*1000)/10,"%.")
```

```python
Simulateur()
```
