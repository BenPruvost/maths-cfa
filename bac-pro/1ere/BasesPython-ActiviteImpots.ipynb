## Les Variables

On a 4 types de variables en Python :
- Entiers int
- Décimaux float
- Booléens bool (True ou False)
- Chaînes de caractères str (pour du texte)

```
une_variable = 10  # une_variable est de type int

une_autre_variable = une_variable/20 # une_autre_variable est un float
print(une_autre_variable)
```

```
# Pour mettre au carré avec la commande **2
une_autre_variable = une_autre_variable**2

# Pour incrémenter une_variable avec +=
une_variable += 1 # On ajoute 1 à une_variable

print(une_variable, une_autre_variable)
```

```
mot = "bonjour"     # mot est une variable str
nom = 'Mr Pruvost'  # on peut utiliser les apostrophes pour des str

# l'opération + sur une str permet de mettre les variables bout à bout
phrase = mot + nom
print(phrase)

phrase = mot + ' ' + nom # il faut ajouter un espace entre mot et nom
print(phrase)
```

## Les Conditions

On peut utiliser les instructions conditionnelles if, elif, else pour introduire du code qui ne se déclenchera qu'en fonction de conditions particulières.

Par exemple :
- Si le revenu est inférieur à 10 000 euros, le taux d'imposition est de 5 %;
- Si le revenu est entre 10 000 et 30 000 euros, le taux d'imposition est de 20 %;
- Si le revenu est supérieur à 30 000 euros, le taux d'imposition est de 40 %.

```
revenu = 5000
taux = 0

# 1ère condition : revenu inférieur à 10 000
if revenu < 10000:
    taux = 0.05 # Attention à bien laisser l'indentation

# 2ème condition : revenu inférieur à 30 000 (on a déjà écarté le cas inférieur à 10 000)
elif revenu < 30000:
    #Attention, la 2ème instruction doit être au même niveau que le premier if
    taux = 0.2
    
# Tous les autres cas (donc revenus supérieurs à 30 000)
else :
    taux = 0.4
    
print("Taux d'imposition :", taux*100,'%')
print()
print("Impot annuel :", revenu*taux,'euros')
```

## Créer une fonction

On peut créer une fonction en utilisant l'instruction def :

```
def TauxImpot(revenu) : #on demande une variable revenu pour utiliser la fonction
    
    #On peut récupérer notre précédent code
    #il faut sélectionner tout le code puis utiliser la touche tab pour bien l'indenter
    
    if revenu < 10000:
        taux = 0.05
    elif revenu < 30000:
        taux = 0.2
    else :
        taux = 0.4
        
    #on veut retourner le taux comme réponse à la fonction
    return taux
```

```
revenu = 5000

#On appelle la fonction pour obtenir le taux d'imposition
taux = TauxImpot(revenu)

print("Taux d'imposition :", taux*100,'%')
print()
print("Impot annuel :", revenu*taux,'euros')
```
